```python
import pandas as pd
```

```python
mvps = pd.read_csv("mvps.csv")
```

```python
mvps
```

Extract only the data that is relevant for MVP voting as we have their player data saved in the other CSV file

```python
mvps = mvps[["Player","Year","Pts Won","Pts Max","Share"]]
```

```python
mvps
```

```python
players = pd.read_csv("players.csv")
```

```python
players
```

```python
del players["Unnamed: 0"]
del players["Rk"]
```

```python
players
```

Next we will remove any * from Player names so that we can properly merge the data with the mvps CSV

```python
players["Player"] = players["Player"].str.replace("*", "", regex=False)
```

```python
players.groupby(["Player","Year"]).get_group(("Greg Anderson", 1991))
```

Players who have multiple teams in one year have multiple rows so we want to only keep the Total stats and keep the last team they were on for that year 

```python
def single_row(df):
    if df.shape[0] == 1:
        return df
    else:
        row = df[df["Tm"] == "TOT"]
        row["Tm"] = df.iloc[-1,:]["Tm"]
        return row

players = players.groupby(["Player","Year"]).apply(single_row)
```

```python
players.head()
```

```python
players.index = players.index.droplevel()
```

```python
players.index = players.index.droplevel()
```

```python
players[players["Player"] == "Greg Anderson"]
```

Now we can merge our two data frames

```python
combined = players.merge(mvps, how="outer", on=["Player", "Year"])
```

```python
combined.head()
```

We must transform the last 3 columns NA values to zeros so that we can use it in out Machine Learning model later

```python
combined[["Pts Won","Pts Max","Share"]] = combined[["Pts Won","Pts Max","Share"]].fillna(0)
```

```python
combined.head()
```

```python
teams = pd.read_csv("teams.csv")
```

We will clean up the original header rows that now say their teams "Division" as a row

```python
teams = teams[~teams["W"].str.contains("Division")]
```

```python
teams["Team"]=teams["Team"].str.replace("*","", regex = False)
```

```python
teams["Team"].unique()
```

```python
combined["Tm"].unique()
```

We will use the uploaded nicknames df in order to create a dictionary with each team abbreviation and it's team name

```python
nicknames = {}

with open("nicknames.csv") as f:
    lines = f.readlines()
    for line in lines[1:]:
        abbrev,name = line.replace("\n","").split(",")
        nicknames[abbrev] = name
```

Now we can use pandas map function to create a full team name column given the nickname dictionary

```python
combined["Team"] = combined["Tm"].map(nicknames)
```

```python
stats = combined.merge(teams, how="outer", on=["Team","Year"])
```

```python
stats
```

```python
del stats["Unnamed: 0"]
```

Next we need to make sure that our columns are of the correct data type so that we can properly use them later on

```python
stats = stats.apply(pd.to_numeric, errors="ignore")
```

```python
stats["GB"].unique()
```

```python
stats["GB"] = stats["GB"].str.replace("—","0")
```

```python
stats["GB"]= pd.to_numeric(stats["GB"])
```

```python
stats.dtypes
```

```python
stats.to_csv("player_mvp_stats.csv")
```

Now that all of our columns are in the proper data types we can explore the data

```python
highest_scoring = stats[stats["G"] > 70].sort_values("PTS", ascending=False).head(10)
```

```python
highest_scoring.plot.bar("Player","PTS")
```
