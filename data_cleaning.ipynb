Extract only the data that is relevant for MVP voting as we have their player data saved in the other CSV file

Next we will remove any * from Player names so that we can properly merge the data with the mvps CSV

```python
players["Player"] = players["Player"].str.replace("*", "", regex=False)
```

```python
players.groupby(["Player","Year"]).get_group(("Greg Anderson", 1991))
```

Players who have multiple teams in one year have multiple rows so we want to only keep the Total stats and keep the last team they were on for that year 

```python
def single_row(df):
    if df.shape[0] == 1:
        return df
    else:
        row = df[df["Tm"] == "TOT"]
        row["Tm"] = df.iloc[-1,:]["Tm"]
        return row

players = players.groupby(["Player","Year"]).apply(single_row)
```

```python
players.head()
```

```python
players.index = players.index.droplevel()
```

```python
players.index = players.index.droplevel()
```

Now we can merge our two data frames

```python
combined = players.merge(mvps, how="outer", on=["Player", "Year"])
```

```python
combined.head()
```

```python
combined[["Pts Won","Pts Max","Share"]] = combined[["Pts Won","Pts Max","Share"]].fillna(0)
```
